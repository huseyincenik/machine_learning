# <p style="background-color:#3498DB; font-family:newtimeroman;color:#FFF9ED; font-size:150%; text-align:center; border-radius: 15px 50px;"> ‚ú® Coefficient of determination (R2) ‚ú®</p>

<button type="button" class="btn btn-lg btn-danger" data-toggle="popover" title="Popover title">Info</button>
<a href="#toc" class="btn btn-warning btn-lg active" role="button" aria-pressed="true" style="color:black;">Content</a>

<div style="background-color:#3498DB; padding:10px; border-radius: 10px; box-shadow: 2px 2px 4px rgba(0, 0, 0, 0.5);">
    <p style="font-weight: bold; color: white;">The R2 score indicates how much of the variation in the target column can be explained by the independent features at hand.</p>
    <p>Statistically, R2 scores typically range from [0, 1], while in the context of Scikit-learn, the range is [-‚àû, 1].</p>
    <p>Seventy-five percent of the dataset's variance can be explained.</p>
</div>


<div style="background-color:#3498DB; padding:10px; border-radius: 10px; box-shadow: 2px 2px 4px rgba(0, 0, 0, 0.5); color: white;">
    <p style="font-weight: bold;"> Example:</p>
    <p>A house (2 + 1) price ----> 50,000 USD </p>
    <p>B house (3 + 1) price ----> 60,000 USD</p>
    <p>If 3,000 USD of the variation in house price is attributable to the number of rooms in the houses, what is the Coefficient of Determination (R2) for the number of rooms?</p>
    <p>R2 Score for the number of rooms in the houses:</p>
    <p>r2_score = 3000 / (60000 - 50000)</p>
    <p>r2_score = 0.3</p>    
</div>

    <p style="font-weight: bold;">LSE stands for least squared errors, while ABS stands for absolute errors.</p>
    <p>It's important to remember that LSE involves squaring the residuals, which means it penalizes high errors, whereas ABS does not apply this squared penalty.</p>
    <p>Residuals represent the errors in the prediction.</p>
    <p>Which metric to use will depend on your specific use case. If your dataset contains many outliers, LSE can be a useful choice.</p>
